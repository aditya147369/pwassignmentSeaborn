Q1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Ans - 1. Histogram (sns.histplot):

Usage - Shows the distribution of a numerical variable. It bins the data into intervals (bins) and shows the frequency of observations within each bin.

Example: Analyzing the distribution of heights in a population.

2. Scatter Plot (sns.scatterplot):

Usage: Displays the relationship between two numerical variables. Each point represents an observation, and the position of the point depends on its values for the two variables.

Example: Exploring the correlation between temperature and ice cream sales.

3. Bar Plot (sns.barplot):

Usage: Shows an estimate of central tendency (usually the mean) for a numerical variable grouped by categories. It is useful for comparing quantities across different categories.

Example: Comparing average scores of students across different subjects.

4. Box Plot (sns.boxplot):

Usage: Summarizes the distribution of a numerical variable across several categories. It shows the median, quartiles, and potential outliers in the data.

Example: Analyzing the distribution of salaries across different job categories.

5. Heatmap (sns.heatmap):

Usage: Visualizes matrix-like data where each cell represents the value of a relationship between two variables. It uses color to represent the magnitude of the relationship.

Example: Displaying correlation matrices or visualizing sales performance across different months.


Q2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

fmri = sns.load_dataset("fmri")

plt.figure(figsize=(12, 6)) 
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")

plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("Signal Change Over Timepoint for Different Events and Regions")
plt.legend(title="Event & Region")

plt.show()

```

Q3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_data = sns.load_dataset("titanic")

plt.figure(figsize=(10, 6))

sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.title('Pclass vs ')
plt.xlabel('Passenger Class')
plt.ylabel('Age')

plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.title('Fare vs Pclass')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')

plt.show()
```

Q4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

diamond = sns.load_dataset("diamonds")
sns.histplot(data=diamond, x='price', hue='cut')
plt.title('Histogram of Diamond Prices by Cut')

plt.show()
```

Q5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
import seaborn as sns

iris = sns.load_dataset("iris")
sns.pairplot(iris, hue='species')
```
